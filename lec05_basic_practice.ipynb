## 1. 파이썬 연산자

```python
x = 5
y = 7
z = 12

# 산술 연산자 순서
print(x + y * z)
print((x + y) * z)
print(x + x**y)

# 비교 연산자 여러 개
print(x < y < z)
print(x + y <= z > (y-x)**2)

# 논리 연산자 여러 개
print(True and True and False)
print((True and True) and False) # 위와 동일

print(True and True and True and False)
print(((True and True) and True) and False) # 위와 동일
```

- 산술 연산자의 순서는 일반적인 사칙연산과 동일한 규칙을 따름. (곱셈, 나눗셈이 먼저)
- 비교 연산자를 여러 개 쓰면 and 조건이 있는 것처럼 취급됨.
- 논리 연산자를 여러 개 이어서 쓰면 왼쪽으로부터 오른쪽으로 묶임.

## 3. 파이썬 제어문
### 3.1. if 문 (조건문)

```python
# 1-line if 문

x = 5
y = 10

x = x if x > 0 else 0 # x가 0보다 클 때는 x를, 작거나 같으면 0을 저장 (아래 4줄의 코드와 완전히 동일)
if x > 0:
  x = x
else:
  x = 0

y = y if y > x else x # y가 x보다 클 때는 y를, 작거나 같으면 x를 저장

print(x, y)
```

```python
# if 문 예제 1 - 두 숫자 중에 더 큰 수를 구하기

x, y = 1, 2

# 일반적인 코드
if x > y:
  max = x
else:
  max = y
print(max)

# 1-line 코드
max = x if x > y else y
print(max)
```

- 1-line으로 표현하는 코드는 파이썬이라는 언어의 가장 큰 강점 중 하나임.
- 간결한 코드를 위해 복잡하지 않은 if 문은 1-line 코드로 설계하는 것을 권장.

```python
# if 문 예제 2 - 둘 중에 더 큰 수를 구하고, 두 수가 같을 때는 0을 출력하게 하기

x, y = 1, 1

if x > y:
  max = x
elif x < y:
  max = y
else:
  max = 0
print(max)

# 1-line version
max = x if x > y else y if x != y else 0 # 1-line 코드에서는 elif 사용 불가능. 대신 중첩된 if-else문 사용.
print(max)
```

### 3.2. for 문 (반복문)

```python
# 1-line for문
numbers = [1,2,5,6,7]

num_plus1 = [num+1 for num in numbers]
print(num_plus1)

num_odd = [num for num in numbers if num%2 == 1] # for, if를 1-line으로 함께 사용 가능
print(num_odd)
```

```python
scores = [10, 50, 40, 75, 60, 99]

# if, for문 예제 1 - 리스트에서 50이 넘는 수는 몇개?
over50 = 0
for score in scores:
  if score > 50:
    over50 += 1
print(over50)

over50 = len([score for score in scores if score > 50]) # 조건문을 충족하는 값만 리스트에 저장한 뒤, 그 길이를 계산하는 방식으로 코드를 간결하게 함
print(over50)
```

```python
# if, for문 예제 2 - 소수 찾기
n = int(input('1부터 N까지의 숫자에서 소수 찾기. N='))
decimals = []

for i in range(2,n):
  for j in range(2,i+1):
    if j == i:
      decimals.append(j)
    if i%j == 0:
      break

print(decimals)
```

### 3.3. while 문 (반복문)

```python
# while 문 예제 1 - 하루 공부했을 때 10점을 올릴 수 있다면, 평균 80점 이상을 달성하기 위해 필요한 시간은? (단, 만점 100점을 넘길 수 없음.)

day = 0
while sum(scores) / len(scores) < 80:
  scores = [score + 10 for score in scores if score + 10 <= 100]
  day += 1

print(day)
```

```python
# while 문 예제 2 - 패스워드 입력받기

passwd = ''
while len(passwd) < 8:
  passwd = input('8자리 이상의 패스워드를 입력하세요. ')
```

## 4. 파이썬 함수

```python
# 구구단 출력하기

for i in range(1,10):
  print('{} * {} = {}'.format(2, i, 2 * i))
for i in range(1,10):
  print('{} * {} = {}'.format(3, i, 3 * i))
for i in range(1,10):
  print('{} * {} = {}'.format(4, i, 4 * i))
for i in range(1,10):
  print('{} * {} = {}'.format(5, i, 5 * i))
for i in range(1,10):
  print('{} * {} = {}'.format(6, i, 6 * i))
for i in range(1,10):
  print('{} * {} = {}'.format(7, i, 7 * i))
for i in range(1,10):
  print('{} * {} = {}'.format(8, i, 8 * i))
for i in range(1,10):
  print('{} * {} = {}'.format(9, i, 9 * i))
```

- 구구단을 출력할 때, 함수를 활용하지 않으면 코드가 길고 비효율적임.

```python
# 함수 예제 1 - 구구단 출력하기
def multiple_table(n):
  for i in range(1,10):
    print('{} * {} = {}'.format(n, i, n * i))

for i in range(2, 10):
  multiple_table(i)
```

- 위와 같이 반복적인 작업을 코드로 설계할 때, 함수를 활용하여 훨씬 효율적이고 짧은 코드를 짤 수 있음.

```python
# 함수 예제 2 - 팩토리얼 계산하기

n = int(input('N! 계산하기. N='))

def factorial(n):
  result = 1
  for i in range(1, n+1):
    result *= i
  return result

print(factorial(n))

# 재귀함수 활용 버전 (난이도 상)
def my_factorial(n):
  if n > 1:
    return n * my_factorial(n-1)
  else:
    return 1

print(my_factorial(n))
```

- 재귀함수 활용의 경우, 다른 컴퓨터 언어로 재귀함수 개념을 접한 적 없을 경우 현재 단계(파이썬 함수 개념 학습)에서는 몰라도 무방함.

```python
# 함수 예제 2.2 - 함수를 활용하여 1부터 10까지 팩토리얼 값 출력

for i in range(1,11):
  print('{}! = {}'.format(i, my_factorial(i)))
```
