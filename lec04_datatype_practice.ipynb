## 2. 변수와 상수
- 한 변수에 다른 변수를 할당했을 때, 즉 상수 자리에 변수를 넣었을 때 일어나는 일

````python
name = 'Taylor'
age = 25
changed_age = 24
age = changed_age

print('Taylor is', age, 'years old.')
````

````python
# 두 변수에 들어있는 값을 서로 바꾸는 방법은?
x = 1
y = 2

x = y
y = x
print(x,y)
````

- x,y라는 변수는 1,2라는 상수가 든 상자이므로 x라는 상자에 들어있던 상수인 2를 넣으면 원래의 값인 1은 사라짐.

````python
# 올바른 답
x = 1
y = 2

z = x
x = y
y = z
print(x,y)
````

- x나 y 둘 중 하나의 값을 임시로 저장해둘 새로운 변수 z가 필요.

## 3. 기본 자료형
### 3.1. 숫자형

````python
x = 1
y = 1.0 # 숫자 1은 정수이지만 1.0으로 정의하면 실수 타입으로 저장
z = x + y

print(x,y,z)
print(type(x),type(y),type(z))
````

### 3.2. 문자형

````python
# 문자열의 덧셈과 곱셈
program_name = 'My Python'
start = program_name + ' starts!' # 덧셈
loading_bar = '='
loading = loading_bar * 100 # 곱셈
end = program_name + "'s finished."

print(start)
print(loading)
print(end)
````

````python
# 줄바꿈 개행문자 \n
whole_program = start + '\n' + loading + '\n' + end
print(whole_program)
````

### 3.3. 논리형

````python
condition1 = True
condition2 = False
condition3 = False

print(condition1 or condition2 and condition3) # 3개 이상의 조건을 걸어 논리 연산자 사용 가능
print(condition1 and condition2)
print(condition1 & condition2) # and 대신 & 사용하기도 함
print(condition1 | condition2) # or 대신 | 사용하기도 함
````

### 3.4. 자료형 변환

````python
# 자료형 변환 응용 문제 1 - int와 bool, str과 bool
zero = 0
not_zero = 1
not_zero2 = 3
print(bool(zero), bool(not_zero), bool(not_zero2))
````

````python
empty_str = ''
not_empty = 'sth'
print(bool(empty_str), bool(not_empty))
````

- boolean 자료형 변환을 통해 int의 경우 0인지 아닌지, str의 경우 문자열이 비어있는지 아닌지 검사할 수 있음.

````python
# 자료형 변환 응용 문제 2 - int와 float
x = 1
y = 1.2

int_y = int(y)
float_x = float(x)

print(int_y, float_x)
````

- 정수에서 실수로 변환 시 소수점 아래 한 자리가 붙어 나옴.
  - 예) 1 -> 1.0
- 실수에서 정수로 변환 시 소수점 아래 값은 버림.
  - 예) 1.2 -> 1

## 4. 시퀀스 자료형
### 4.1. String

````python
py_str = 'How To Index and Slice in Python'
print(len(py_str)) # 문자열 길이 측정
````


```
str[start:end:step]
```
- start는 시작 인덱스, end는 끝 인덱스-1, step은 건너뛸 인덱스 개수
- 비워둘 경우 0으로 간주됨.


````python
print(py_str[:]) # 전부 출력
print(py_str[-6:]) # 뒤에서 6번째부터 끝까지 출력
print(py_str[::4]) # 처음부터 끝까지 4개씩 건너뛰며 출력
print(py_str[:26:7]) # 처음부터 25번째까지 7개씩 건너뛰며 출력
print(py_str[5:15:2]) # 5번째부터 14번째까지 2개씩 건너뛰며 출력
print(py_str[::-1]) # 처음부터 끝까지 뒤로 가며 거꾸로 출력
````

````python
print('Python' not in py_str) # not in 사용 가능
print('HOW TO' in py_str)
print('How To' in py_str) # 대소문자 구분
````

### 4.2. Tuple

````python
py_tuple = (1,) # 튜플 값이 하나일 때는 반드시 뒤에 콤마(,) 추가
py_tuple = (1, 2, 'one', 'two')
print(py_tuple[:2]) # 인덱싱, 슬라이싱 가능
````

### 4.3. List

````python
py_list = [1, 2, 'one', 'two']

py_list[0] = 2 # 값 수정
print(py_list)

py_list.append('three') # 추가
print(py_list)

py_list.remove('three') # 값을 찾아서 삭제
print(py_list)

del py_list[0] # 인덱스로 삭제
print(py_list)
````

````python
# 두 리스트 합치기
py_list2 = [3, 'three']
py_list3 = py_list + py_list2 # 더하기(+) 연산자로 합치기
print(py_list3)
py_list.extend(py_list2) # extend 함수로 py_list 뒤에 py_list2 붙이기
print(py_list)

py_list.append(py_list2) # 주의) append 함수 사용 시 두 리스트가 합쳐지는 게 아니라 리스트 안에 리스트가 들어감.
print(py_list)
````

### 4.4. Dictionary

````python
py_dict = {'one': 1, 'two': 2, 'three': 3}
py_dict['four'] = 4 # 새로운 key, value 추가하기
print(py_dict)
print(py_dict['one']) # key 값으로 value 찾기 1
print(py_dict.get('one')) # key 값으로 value 찾기 2
````

- 주의) 모든 딕셔너리의 key는 중복 불가

````python
# key 값으로 value 찾는 방식 2가지의 차이점 (없는 key 값을 불러오려고 시도할 때)
print(py_dict.get('five')) # None 반환
````

````python
print(py_dict['five']) # 에러 발생
````

````python
print(py_dict.keys()) # 딕셔너리의 key 리스트
print(py_dict.values()) # 딕셔너리의 value 리스트
````

### 4.5. Set

````python
py_list = [1,1,2,3,3,4,5]
py_set = set(py_list) # set은 중복을 허락하지 않는 자료형이므로 변환을 통해 리스트의 중복 제거 가능
print(py_list)
print(py_set)
````
